## Decision Tree (의사 결정 트리)

* 어떤 속성을 적용해서 분리를 할지를 결정
	* 변별력이 좋은 질문 = 분리 결과의 엔트로피가 낮은 속성을 선택
	
<br/>

* 엔트로피 : 무질서도 (높으면 그만큼 무질서함), 0~1 사이의 값을 가짐
	* $ E(s) = \displaystyle\sum^{c}_{i=1} -p_i \log_2 p_i$

```python
# sklearn을 이용한 의사 결정 트리
from sklearn import tree
X = [[0, 0], [1,1]]
Y = [0, 1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)

clf.predict([[2, 2]])
```

### 의사 결정 트리를 이용한 붓꽃 분류

```python
# 라이브러리 호출
import numpy as np
from sklearn import datasets, metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
# 붓꽃 데이터 가져오기
iris = datasets.load_iris()
X, y = iris.data, iris.target
```

```python
# 학습 데이터 분리 (80 : 20)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50, test_size=0.2)
```

```python
# 의사 결정 트리 모델 생성 & 학습
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_train)
print("학습 데이터 정확도 :", accuracy_score(y_true=y_train, y_pred=y_pred))
```

```python
y_pred = clf.predict(X_test)
print("테스트 데이터 정확도 :", accuracy_score(y_true=y_test, y_pred=y_pred))
```

```python
# plot_tree()를 이용하여 트리 플롯
from sklearn.tree import plot_tree

plot_tree(clf)
```
