```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
import numpy as np
```

## K-means Clustering

k개의 클러스터로 분할하는 알고리즘 -> `비지도 학습법`

클러스터의 중심을 옮겨가면서 최적의 위치를 찾는 휴리스틱 알고리즘으로 구현

```python
# 데이터 준비
X = np.array([
	[6,3], [11,15], [17,12], [24,10], [20,25], [22,30],
	[85,70], [71,81], [60,79], [56,52], [81,91], [80,81]
])

# 데이터 시각화
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# 클러스터가 2개인 K-means 알고리즘
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
# 시각화
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap="rainbow")
plt.show()
```

## 클러스터의 개수 정하기

올바른 클러스터의 수를 정해야 함 -> `팔꿈치 방법(elbow method)`으로 결정

k를 1부터 증가시키면서, 각 k의 값에 대해서 SSE(Sum of Squared Errors) 값을 계산

```python
n_clusters = range(1, 10)
kmeans = [KMeans(n_clusters=i, n_init=10) for i in n_clusters]

score = [kmeans[i].fit(X).inertia_ for i in range(len(kmeans))]

plt.plot(n_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```
